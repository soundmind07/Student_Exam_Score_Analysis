```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv(r"C:\Users\Pallavit\Downloads\archive (6)\Expanded_data_with_more_features.csv")
df
```

```python
df.isnull().sum()
```

```python
df=df.drop("Unnamed: 0",axis=1)
df.head()
```

```python
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distrubition")
plt.show()
```

From the above chart we analysed that number of females is more than number of males.

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Relationship between parents and student score")
plt.show()
```

from the above chart we analysed that education of parents has a good impact on students score

```python
gb=df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"})
sns.heatmap(gb,annot=True)
plt.title("Relationship between parents marital status and student score")
plt.show()
```

from here we analysed that their is negligible impact of their marital status on student score.

```python
groupA=df.loc[(df["EthnicGroup"]=="group A")].count()
groupB=df.loc[(df["EthnicGroup"]=="group B")].count()
groupC=df.loc[(df["EthnicGroup"]=="group C")].count()
groupD=df.loc[(df["EthnicGroup"]=="group D")].count()
groupE=df.loc[(df["EthnicGroup"]=="group E")].count()
l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("Disstribution of Ethnic Groups")
plt.show()
```
